<h1>  Bar Chart </h1>


```python
categories=["Grains","Fruit","Veg","protin"]
values=np.array([4,3,2,5])

plt.bar(categories, values , color="black")

# for vertical barc Chart
#plt.barh(categories, values)

plt.title("Daily Consumption")
plt.xlabel("Food" , color="red")

plt.show()
```

<h1> Pie Chart </h1>

```python
categories=np.array(["Grains","Fruit","Veg","protin"])
values=np.array([4,3,2,5])
colors=["red","yellow","blue","green"]

plt.pie(values , labels=categories ,autopct="%1.1f%%" , colors=colors , explode=[0,0,0,0.15])

plt.title("VGEC" , color="darkblue")
```

<h1> scatter plot</h1>


--> Shows the relationship between two variables helps to identify a correlation (positive , negative , none) 

example - Study Hours vs Test scores


```python
x1=np.array([0,1,2,3,4,5,6,7])
y1=np.array([55,60,65,70,75,80,85 ,90])


x2=np.array([0,1,2,3,4,5,6,7])
y2=np.array([50,55,70,65,70,60,80,85])

plt.scatter(x1,y1, color="purple", s=200 ,  label="classA")  # s-size

plt.scatter(x2,y2, color="red", s=200 , label="classB")  



plt.legend()
plt.title("Test Score")
plt.xlabel("Hours Studied")

```

<h1>Histogram</h1>


--> A visula representation of the distribution of quantitative date. 
--> They grop balues into bins (intervals).
--> and counts how many fall in each range

```python
scores=np.random.normal(loc=80, scale=10, size=100)

scores=np.clip(scores,0,100)

plt.hist(scores,bins=10 ,color="lightgreen", edgecolor="black")
plt.savefig('scatterplot.svg')
```

```python

x=np.array([1,2,3,4])
figure,axes=plt.subplots(2,2)

axes[0,0].plot(x , x*2)
axes[0,0].set_title("x*2")

axes[0,1].bar(x , x**2 )
axes[0,1].set_title("x**2")


axes[1,0].plot(x , x**3, color="purple")
axes[1,0].set_title("x**3")

axes[1,1].plot(x , x**4)
axes[1,1].set_title("x**4")


plt.tight_layout()

```
