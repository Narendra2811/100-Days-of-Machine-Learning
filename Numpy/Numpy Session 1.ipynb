<h1> Array Function</h1>

```python
a1=np.random.random((3,4))
a1=np.round(a1*100)
a1
```

```python
#max/min/prod
# axis =0,1 --> 0-column ,1- row
np.sum(a1,axis=1)
```

```python
np.max(a1,axis=0)
```

```python
# mean/median/std/var
np.var(a1)
np.mean(a1,axis=0)
```

```python
# dot product
a2=np.arange(12).reshape(3,4)
a3=np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
# log and exponents

np.log(a1)
np.exp(a1)
```

```python
#random/floor/cell
#ceil (2,7 ->2)
#floor (3.1 -->4)

np.round (np.random.random((2,3))*100)
```

# Stacking

```python
a4=np.arange(12).reshape(3,4)
a5=np.arange(12,24).reshape(3,4)

```

```python
np.hstack((a4,a5,a4))
```

```python
np.vstack((a4,a5))
```
